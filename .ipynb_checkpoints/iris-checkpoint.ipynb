```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Convert categorical 'species' to numerical values
iris['species_num'], unique = pd.factorize(iris['species'])
```

```python
iris.head()
```

```python
iris.corr(numeric_only=True)
```

```python
sns.heatmap(iris.corr(numeric_only=True))
```

```python
# sns.jointplot(x='petal_length', y='petal_width', data='iris')
sns.jointplot(x='petal_length', y='species_num', data=iris, kind='reg' )
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris , hue='species_num')
```

```python
sns.distplot(iris['species_num'])
```

```python
sns.countplot(x='petal_length', data=iris)
```

```python
sns.barplot(x='petal_length', y='sepal_length', data=iris, hue='species_num', width=0.5, gap=2)
```

```python
sns.boxplot(y='sepal_length', x='species_num', data=iris, palette = 'dark')
```

```python
sns.boxplot(y='sepal_length', x='species_num', data=iris, palette = sns.cubehelix_palette(n_colors=8, reverse=False ) , saturation=0.75)
```

```python
sns.violinplot(x="species_num", y="petal_width", data=iris,palette='rainbow')
```
